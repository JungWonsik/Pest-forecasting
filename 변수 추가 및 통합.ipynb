## 유아등추가

```python
import pandas as pd
dat = pd.read_csv('C:/Users/user/Desktop/project/bug_weather_toyang\
/해충 병합 데이터/merge/merge_2018.csv',encoding='ms949')
data1=dat[dat['시도명']=='전라남도']
df=data1[data1['병해충명']=='먹노린재(유아등)'].sort_values\
(by=['입력기준일자','병해충조사지점번호','조사회차'])
df1 = df[['입력기준일자','시군구명','조회값']]
df2 = df1.groupby(['입력기준일자','시군구명']).sum()
data2018=pd.merge(data,df2,on=['입력기준일자','시군구명'],how='left')
data2018=data2018.rename(columns={'조회값':'유아등'})
```

# 20주당 마리수 추가

```python
dat1 = pd.read_csv('C:/Users/user/Desktop/project/bug_weather_toyang\
/해충 병합 데이터/merge/merge_2017.csv',encoding='ms949')
dat2=dat1[dat1['시도명']=='전라남도']
a=dat2[dat2['병해충명']=='먹노린재(20주당마리수)']
a1 = a[['시군구명','조회값']]
a2 = a1.groupby('시군구명').sum()
data2018 = pd.merge(data2018,a2,on='시군구명',how='left')
data2018=data2018.rename(columns={'조회값':'지난해출현수'})
```
