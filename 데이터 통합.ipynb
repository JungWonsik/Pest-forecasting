```python
data1=data[data['시도명']=='전라남도']
```

```python
data2=data1[data1['병해충명']=='먹노린재(감수율)'].sort_values(by=['입력기준일자','병해충조사지점번호','조사회차'])
data3=data1[data1['병해충명']=='먹노린재(발생면적)'].sort_values(by=['입력기준일자','병해충조사지점번호','조사회차'])
data4=data1[data1['병해충명']=='먹노린재(발생면적률)'].sort_values(by=['입력기준일자','병해충조사지점번호','조사회차'])
data5=data1[data1['병해충명']=='먹노린재(피해율)'].sort_values(by=['입력기준일자','병해충조사지점번호','조사회차'])
df=data1[data1['병해충명']=='먹노린재(유아등)'].sort_values(by=['입력기준일자','병해충조사지점번호','조사회차'])
```

```python
data6=pd.merge(data2,data3[['입력기준일자','병해충조사지점번호','조사회차','조회값']],on=['입력기준일자','병해충조사지점번호','조사회차'])
data6=data6.rename(columns={'조회값_x':'감수율','조회값_y':'발생면적'})
data6=pd.merge(data6,data4[['입력기준일자','병해충조사지점번호','조사회차','조회값']],on=['입력기준일자','병해충조사지점번호','조사회차'])
data6=pd.merge(data6,data5[['입력기준일자','병해충조사지점번호','조사회차','조회값']],on=['입력기준일자','병해충조사지점번호','조사회차'])
data6=data6.rename(columns={'조회값_x':'발생면적률','조회값_y':'피해율'})
data6
```
